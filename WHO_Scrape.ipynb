```python
#Importing The Necessary Libraries.

import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#The URL of the required site to scrape

url ='https://www.who.int/health-topics'
```

```python
#Accessing the url with our request.get function

response = requests.get(url)
```

```python
#Checking The Response to know if its successcful(200 means successful)

print((f'{response.status_code}; means response is succesful'))
```

```python


soup = BeautifulSoup(response.text,'html.parser')
```

```python
soup
```

```python
#To Get Names and Links off the site

topics = []

for link in soup.find_all('a', class_="link-container table", href=True):
    topic_name = link.find('p', class_="heading text-underline").get_text()
    topic_link = link['href']
    topics.append({'name':topic_name, 'link':topic_link})
```

```python
#collecting Data for all topics

data = []

for topic in topics:
    print(f'Scraping Content: {topic['name']}')
    #Fetching each page html
    topic_response = requests.get(topic['link'])
    topic_soup = BeautifulSoup(topic_response.text, 'html.parser')

    #Locate all content divs with the specific class
    content_divs = topic_soup.find_all('div', class_="sf_colsOut tabContent")

    #creating a dictionary to contain all content of each topic
    content = {
        'Overview':'N/A',
        'Impact':'N/A',
        'WHO Responses':'N/A'
    }

    #Checking for the availability of each content
    if len(content_divs) > 0:
        content['Overview'] = content_divs[0].get_text(strip= True, separator= '\n')
    if len(content_divs) > 1:
        content['Impact'] = content_divs[1].get_text(strip= True, separator= '\n')
    if len(content_divs) > 2:
        content['WHO Responses'] = content_divs[2].get_text(strip= True, separator= '\n')

    #Append each content to the data
    data.append({
        'Health Topics':topic['name'],
        'Links':topic['link'],
        'Overview':content['Overview'],
        'Impacts':content['Impact'],
        'WHO Responses':content['WHO Responses'],
    })


#creating a dataframe
df = pd.DataFrame(data)

    

```
